```python
from IPython.display import Image

class ShoppingCart:
    def __init__(self):
        self.items = {}

    def add_item(self, item_number, quantity, item_name, price):
        if item_name in self.items:
            self.items[item_name]['quantity'] += quantity
        else:
            self.items[item_name] = {'quantity': quantity, 'price': price}
        print(f"\n{quantity} {item_name} Has Been Added To Your Cart.")

    def remove_item(self, item_name):
        if item_name in self.items:
            del self.items[item_name]
            print(f"\n{item_name} Was Removed From Your Cart.")
        else:
            print(f"\n{item_name} Not Found in Your Cart.")

    def view_cart(self):
        if not self.items:
            print("\nNo Items in Cart.")
        else:
            print("\nItems in Cart:")
            total_cost = 0
            for item, details in self.items.items():
                quantity = details['quantity']
                print(f"{item}: {quantity} - ${details['price']}")
                cost = details['price'] * quantity
                total_cost += cost
            print(f"\nCart Total: ${total_cost}")

class AccelerationIndustriesStore:
    def __init__(self):
        self.items = {
            1: {'name': 'Borla Exhaust Kit', 'price': 560, 'image_path': 'borla_exhaust.jpg'},
            2: {'name': 'Big Turbo Upgrade Kit', 'price': 2000, 'image_path': 'big_turbo_upgrade.jpg'},
            3: {'name': 'Wheel and Tire Combo 20 Inches (Set of 4)', 'price': 1000, 'image_path': 'gold_rims.jpg'},
            4: {'name': 'GT Style Wing 70 Inches (Universal Brackets)', 'price': 535, 'image_path': 'gt_wing.jpg'},
            5: {'name': 'Intercooler Upgrade', 'price': 712, 'image_path': 'intercooler_upgrade.jpg'},
            6: {'name': 'Tuning Chip Upgrade (Racing Chip US)', 'price': 528, 'image_path': 'race_chip.jpg'},
            7: {'name': 'Widebody Kit (Chilling Tuning)', 'price': 1450, 'image_path': 'widebody_kit.jpg'}
        }
        self.cart = ShoppingCart()

    def display_store_menu(self):
        print("\nACCELERATION INDUSTRIES Store\n")
        for item_number, item_info in self.items.items():
            image_path = item_info['image_path']
            display(Image(filename=image_path, embed=True))
            print(f"{item_number}. {item_info['name']} - ${item_info['price']}")

    def display_cart_menu(self):
        print("\nCART MENU\n")
        print("1. Add Item")
        print("2. Delete Item")
        print("3. View Cart")
        print("4. Exit")

    def start_shopping(self):
        while True:
            self.display_store_menu()
            self.display_cart_menu()

            option = input("\nEnter Option 1-4: ")

            if option == '1':
                self.display_store_menu()
                item_number = int(input("Enter Item Number to Add to Cart: "))
                if item_number in self.items:
                    quantity = int(input("Enter Item Quantity: "))
                    item_name = self.items[item_number]['name']
                    price = self.items[item_number]['price']
                    self.cart.add_item(item_number, quantity, item_name, price)
                else:
                    print("Item Number Not Found. Choose a Valid Item Number.")

            elif option == '2':
                item_to_remove = input("\nEnter Item Name to Delete From Cart: ")
                self.cart.remove_item(item_to_remove)

            elif option == '3':
                self.cart.view_cart()

            elif option == '4':
                self.cart.view_cart()
                print("\nThank You for Shopping at ACCELERATION INDUSTRIES!")
                break

            else:
                print("Invalid Choice. Enter a Choice from 1 to 4.")

if __name__ == "__main__":
    store = AccelerationIndustriesStore()
    store.start_shopping()
```
